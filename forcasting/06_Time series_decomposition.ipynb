## 6.1 Time series components

#### Additive decomposition

* pattern의 합으로 분해 
* seasonal component나 trend-cyclic component가 시간의 흐름에 따라서 변화하지 않는 경우(does not vary on the level of time)

![](./images/ch06/01.png)

#### Multiplicative decomposition

* 곱으로 분해
* seasonal component나 trend-cyclic component가 시간의 흐름에 따라서 변화화는 경우(does vary on the level of time)
* economic data는 주로 여기에 해당 

![](./images/ch06/02.png)

#### decompositipon with transformation

* transformation을 통해서 variation이 시간이 지남에 따라서 변하기 않게 한 후에 additive decomposition 적용
* 만역 log-transform을 사용했다면 addtive == multiplicative

![](./images/ch06/03.png)

###  Seasonally adjusted data

* seasonal component를 원본 data에서 제거한 것
  * y_t - S_t  or y_t / S_t
  
![](./images/ch06/06.png)

* 당연히 seasonal component가 주요 관심사가 아닌 경우에 이용된다. 
* ex) 실업률 = 계절적 요인(ex. 대학교 졸업 시즌) + 비계절적 요인(ex. 경기 후퇴/하강)
  * 경제학자들은 보통 비계절적 요인에 의한 실업률 추이를 알고 싶으므로 계절적 요인을 제거한 데이터를 얻고 싶어한다. 
* 제거후에도 여전히 remainder component가 있어서 smooth 하지 않으므로 지역적 해석에는 주의

## 6.2 Moving averages

Trend-cyclic component를 구하는 전형적인 방법


#### What and Why?
  * window를 움직여서 local time series에 대해서 평균한 것
  * data가 smooth하다면, 이웃 데이터와의 값 차이가 작고, 이를 평균하면, 데이터의 randomness는 제거하여 smooth trend-cycle을 얻을 수 있다는 가정

#### How?
  * periodic of m 데이터에 대해.. ex) quartely data = m = 4
  * m = 2k+1
  * m-MA 라고 명명
![](./images/ch06/07.png)

#### 예시

* m이 홀수(3,5,7)이면 symmetric하지만, 짝수이면 symmetric하지 않다. 
  * window냐의 중심 데이터에 대해 양쪽의 개수가 차이가 있다.

![](./images/ch06/08.png)
![](./images/ch06/09.png)

### Moving averages of moving averages

* MA를 두번 하는 것
* m이 짝수인 경우에 생기는 non-symmetric을 해소하고자 하는 것
* ex) 2x4-MA
  * 4-MA를 통해 얻은 series에 다시 2-MA를 적용한 것
  * symmetric하게 바뀌었다 (중심 data 좌우의 개수가 같다)


![](./images/ch06/10.png)


### Estimating the trend-cycle with seasonal data

* seasnaol한 데이터에 MA를 잘 적용하는 방법론을 설명
* 2x4-MA는 querterly seasnoal 데이터에 대해서 trend-cyclic component를 구하는 좋은 방법일 수 있다. 
  * 왜냐하면, 일단 m=4로 확실히 추정되고, 짝수인 경우 symmetricity 형성을 위해서 2x4-MA를 하는 것이 추전된다. 
  * 아래 식에서 양 극단을 제외하고는 모든 seasonal data point에 대해서 같은 weight로 감쇠한 후 평균을 하므로, 계절적 요인이 제거될 것이 예상
  
![](./images/ch06/11.png)


* 하지만, 2x8-MA나 2x16-MA도 비슷하게 좋은 결과를 내기는 한다. 

* Rule of thumb 
  * seasonality m을 추정한다. 
  * m이 짝수라면, 2xm-MA를 한다. m이 홀수라면 가냥 m-MA를 한다.
  * ex) monthly data => 2x12-MA, 
  * ex) weekly data => 7-MA
  
### Example: Electrical equipment manufacturing

* monthly data이므로 2x12-MA 적용

## 6.3 Classical decomposition

1920년대-50년대 유행했던 기법

### How-to?

* 한 마디로, MA로 trend-cyclic을 구하고, 이를 이용해서 seasonal 을 구하고, 이를 이용해서 remainder를 구한다. 

 ![](./images/ch06/14.png)
 
### ex) 유럽 설비 데이터를 multiplicative decomposition하기 

* monthly data이라서, 2x12-MA로 trend-cyclic을 구하고, 이를 나누어서 detrended data를 얻음
* detrended data에서 월별로 데이터를 모아서 월별로 평균을 냄, 이를 repeat해서 S_t를 얻음
* 원본 데이터에서 T_t, S_t를 나누어서 remainder를 얻음
 
 ![](./images/ch06/15.png)
 
### 비판 

* MA의 특성상 원본 데이터 양 끝쪽의 estimate가 없음, 특히 MA of MA는 없는 갯수가 더 늘어남 
  * 이에 따라 T_t, S_t, R_t의 추정 개수가 줄어듬
* The trend-cycle estimate tends to over-smooth rapid rises and falls in the data (?)
* Seasonal component는 year by year 동일하게 정말 상수처럼 반복된다고 보지만, 실제는 그렇지 않은 경우가 많다. 
  * 예를 들어 전기 수요 데이터를 다룬다면, 과거에서는 난방때문에 겨울철에 seasonal strength가 높았지만, 요즘은 에어콘 때문에 여름철의 strength가 더 크다. 
* 이따금씩의 outlier 와 같은 seasonal pattern이 있다. 
  * 예를 들어 항공기 여객수 데이터에서 대한항공 사태와 같은 industrial dispute가 급작스럽게 발생해서 생기는 일시적인 seasonal pattern

## 6.7 Measuring strength of trend and seasonality

[Wang, X., Smith, K. A., & Hyndman, R. J. (2006). Characteristic-based clustering for time series data. Data Mining and Knowledge Discovery, 13(3), 335–364](https://robjhyndman.com/publications/ts-clustering/)

### Measuring strength of trend

* Additive decomposition 에 대해서
  * Y(t) = T(t) + S(t) + R(t)
  
* 먼저 S(t)를 구하고, 이를 제거하여 seasonal adjusted data = T(t) + R(t)를 얻는다. 
* 만약 T(t)가 강력하다면
  * R(t)에 의한 variance 대비 T(t) + R(t)를 합한 데이터의 variance는 더 클 것이다.  
  * => 비율은 0
* 만약 T(t)가 약하다면, 
  * R(t)에 의한 variance 대비 T(t) + R(t)를 합한 데이터의 variance는 거의 비슷할 것이다.  
  * => 비율은 1
