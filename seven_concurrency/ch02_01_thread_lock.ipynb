# Ch2, 1일차 : 스레드와 잠금장치

스레드와 잠금장치
- 수많은 문제점에도 불구하고 많은 동시성 소프트웨어를 작성할 때 기본 선택
- 실제 하드웨어가 동작하는 방식을 그대로 옮긴 것. 엄청난 장점임과 동시에 엄청난 단점
- 너무 단순해서 대부분의 언어가 어떤 형태로든 지원
- 쉽게 쓸수는 있으나 부작용을 커버할 기제가 없으며 많은 유지보수 문제사항을 유발

모범 사례
- 낮은 수준의 스레드와 잠금장치를 쓰기 보다는
- 상위 수준을 사용하되, 
- 그 바탕이 되는 낮은 수준을 철저히 이해해야 한다.


## 상호 배제와 메모리 모델 

상호 배제
- 잠금장치를 사용해서 특정 데이터에 접근하는 스레드가 한번에 하나만 존재하도록 강제
- 부작용으로 경쟁조건(race condition)이나 데드락이 발생한다.

메모리 미스테리
- 코드 수행 순서가 바뀔 수 있다.
 - by 컴파일러, 정적 최적화
 - by JVM, 동적 최적화
 - by HW, 파이프라인
- 어떤 스레드의 동작 결과가 다른 스레드에게 보이지 않을 때도 있다. 

메모리 가시성과 자바 메모리 모델
- 자바 메모리 모델은 한 스레드가 메모리에 가한 변화가 다른 스레드에 보이는 경우를 정의
- 읽는 스레드와 쓰는 스레드가 동기화되어야 가시성이 보장된다. 
 - ex) getCount()나 increment() 모두 synchrnized로 선언되어야 한다.
 
데드락
- 하나의 스레드에서 여러 개의 공유자원을 잠금 장치로 확보하려고 할때 발생
- 식사하는 철학자 
- 피하는 방법은 공유자원을 요청할 때 미리 정해진 공통의 순서를 따르면 된다.
 - ex) 자원마다 id부여, 
 - hash로 하면 일반적으로는 안되겠으나 부득이한 경우는 이렇게라도 해야

외부 메서드의 위험
- 잠금을 건 후에 외부 메소드를 호출하는 것은 위험. 
 - ex) lister calling 패턴
 - 왜냐하면 그 메소드 안에서 다른 공유자원 잠금을 시도할 수도 있고, 이 경우 통제가 안되는 데드락 상황 발생 가능
- 대안은 잠금을 걸고 copy로 listeners들을 복사하고, 잠금을 푼 상태에서 listener callback을 수행

요약
- 공유되는 변수에 대한 접근은 반드시 동기화한다.
- 쓰는 스레드와 읽는 스레드가 모두 동기화되어야 한다.
- 여러 개의 잠금장치를 미리 정해진 공통의 순서에 따라 요청한다.
- 잠금장치를 가진 상태에서 외부 메소드를 호출하지 않는다.
- 잠금장치는 최대한 짧게 보유한다.

자율학습
- Double-checking locking
 - singleton 객체를 한번만 생성하도록 보장하는 기법
 - 다만 1.4 에서는 스레드간 가시성이 보장이 안되어서 anti-pattern이었고 
 - 1.5에서는 volatile을 쓴다면 가시성이 보장되서 쓸 수는 있음 
 - 패턴 자체는 anti-pattern이므로 언어마다의 특성을 고려해서 써야 하겠음
